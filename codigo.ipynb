```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats
```

```python
data=pd.read_csv('vistara.csv')
```

```python
df=sns.set(font_scale=1)
df=plt.figure(figsize=(10, 10)) 
df=sns.barplot(data=data, x='departure_time', y="price")
df=plt.xlabel('Tiempo llegada')
df=plt.ylabel('Precio')
df=plt.title('Distribución de los datos de arrivo')
df=plt.xticks(rotation=90)
df=plt.show()
```

```python
df=sns.set(font_scale=1)
df=plt.figure(figsize=(10, 10)) 
df=sns.barplot(data=data, x='class', y= "price")
df=plt.xlabel('class')
df=plt.ylabel('Count')
df=plt.title('Distribución cuidades de destino')
df=plt.xticks(rotation=90)
df=plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='duration', bins=50, kde=True)  # Ajusta el número de bins según tus necesidades
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.title('Distribution of Duration')
plt.show()
```

```python
df=sns.set(font_scale=1)
df=plt.figure(figsize=(10, 10)) 
df=sns.barplot(data=data, x='days_left', y='price')
df=plt.xlabel('days_left')
df=plt.ylabel('price')
df=plt.title('Días de salida vs Precio')
df=plt.xticks(rotation=90)
df=plt.show()
```
